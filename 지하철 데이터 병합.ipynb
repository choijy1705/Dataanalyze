## 여러개의 파일 하나로 합치기

```
raw = pd.DataFrame()

```

```
import os

files = os.listdir('/content/drive/My Drive/Colab Notebooks/data/rawfiles')
files
```

```
# 파일 하나씩 불러와서 한파일로 합치기

for file in files:

  fpath = '/content/drive/My Drive/Colab Notebooks/data/rawfiles/' + file
  # print(fpath)

  temp = pd.read_csv(fpath)
  raw = raw.append(temp, ignore_index=True) #기존 인덱스는 무시하고 병합
```

### 요일 추가하기

```
from datetime import datetime
```

```
# 날짜 컬럼을 불러와서, 순서대로 요일을 점검하여 리스트에 저장

weekday_dict = [ '월','화','수','목','금','토','일']
weekday_list = []

for date_str in raw['사용일자']:

  date = datetime.strptime(str(date_str), '%Y%m%d')
  weekday_index = date.weekday()
  weekday = weekday_dict[weekday_index]
  weekday_list.append(weekday)




```

```
raw['요일'] = weekday_list

raw.head()
```

```
raw.columns
```

```
# 요일을 사용일자 다음에 나오도록 컬럼 순서를 변경하겠습니다. 

raw = raw[['사용일자', '요일', '노선명', '역ID', '역명', '승차총승객수', '하차총승객수', '등록일자']]
```
